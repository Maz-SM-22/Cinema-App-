```python
import requests
import re
from lxml.html import fromstring
import json 
import os
```

```python
url = 'https://www.bolichecinemes.cat/cartelera'
base_url = url[0:31]
```

```python
class Film_Scraper: 
    def __init__(self, url): 
        self.url = url 
    def get_html(self): 
        try: 
            response = requests.get(self.url)
            response.raise_for_status()
        except requests.RequestException: 
            print('A problem occurred with your request')
            raise
        html = fromstring(response.content)
        return html
    def get_film_info(self):
        page = self.get_html()
        screenings = []
        film = {}        
        times = page.xpath('.//a[contains(@href, "https://boliche.admit-one.eu")]/@title')
        for time in times: 
            print(time)
            film['link'] = page.xpath('.//a[@title="{}"]/@href'.format(time))[0]
            print(film['link'])
            datetime = time 
            film['time'] = datetime.split()[-1]
            film['date'] = '{}-{}-{}'.format(datetime.split()[0][0:4], datetime.split()[0][4:6], datetime.split()[0][6:8])
            film['film'] = {}
            try: 
                film['film']['title'] = page.xpath('.//a[@title="{}"]/ancestor::article//h2/a/@data-tiulo')[0]
            except: 
                continue
            film['film']['poster'] = page.xpath('.//a[@title="{}"]/ancestor::article//figure/img/@src')[0]
            film_info = page.xpath('.//a[@title="{}"]/ancestor::article//table')[0]
#             film['film']['year'] = film_info.xpath('.//th[text()="FECHA ESTRENO:"]/following-sibling::td/text()')[0][-4:]
#             film['film']['duration'] = film_info.xpath('.//th[text()="DURACIÓN:"]/following-sibling::td/text()')[0] + 'mins'
#             film['film']['director'] = film_info.xpath('.//th[text()="DIRECTOR:"]/following-sibling::td/text()')[0]
#             film['film']['cast'] = film_info.xpath('.//th[text()="REPARTO:"]/following-sibling::td/text()')[0]
#             film['film']['original_language'] = 'English'
#             film['film']['language'] = 'VOSE'
            film['cinema'] = {}
            film['cinema']['name'] = 'Boliche Cinemes'
            film['cinema']['barri'] = 'Sant Martí'
            film['cinema']['address'] = 'Avinguda Diagonal, 508, 08006 Barcelona, Spain'
            screenings.append(film)
        return screenings
```

```python
scraper = Film_Scraper(url)
results = scraper.get_film_info()
```
