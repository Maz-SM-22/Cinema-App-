```python
import requests
import re
from lxml.html import fromstring
import json
import os
```

```python
url = 'https://barcelona.cines-verdi.com/cartelera'
```

```python
class Scraper: 
    def __init__(self, url): 
        self.url = url 
    def get_html(self): 
        try: 
            response = requests.get(self.url)
            response.raise_for_status()
        except requests.RequestException: 
            print('A problem occurred with your request')
            raise
        html = fromstring(response.content)
        return html
    def get_all_films(self): 
        results = {}
        page = self.get_html()
        page_results = page.findall('.//article[@class="row article-cartelera"]')    
        for item in page_results: 
            item_title = item.xpath('.//h2/a/@title')[0]
            item_url = 'https://barcelona.cines-verdi.com' + item.xpath('.//h2/a/@href')[0]
            results[item_title] = item_url 
        return results
```

```python
scraper = Scraper(url)
all_films = scraper.get_all_films()
```

```python
class Film_Scraper(Scraper): 
    def __init__(self, url): 
        super().__init__(url)
    def get_film_info(self):
        screenings = []
        film = {}                   # Add genre 
        page = self.get_html()
        times = page.findall('.//div[@aria-labelledby="home-tab"]//a')
        for time in times: 
            film['link'] = time.xpath('.//@href')[0]
            film['time'] = time.xpath('.//text()')[0].strip()
            date = page.xpath('.//a[@href="{}"]/ancestor::div[@aria-labelledby="home-tab"]/@id'.format(film['link']))[0][-8:]
            film['date'] = '{}-{}-{}'.format(date[0:4], date[4:6], date[6:8])
            film['film'] = {}
            film['film']['title'] = page.xpath('.//h1/text()')[0].strip()
            film['film']['poster'] = page.xpath('.//figure[contains(@class, "poster")]/img/@src')[0]
            film['film']['year'] = page.xpath('.//div[@class="release-date"]/text()')[0][6:]
            film['film']['genre'] = page.xpath('.//th/following-sibling::td/a/text()')[0]
            film['film']['duration'] = page.xpath('.//th[text()="DURADA:"]/following-sibling::td/text()')[0] + 'min'
            film['film']['director'] = page.xpath('.//th[text()="DIRECTOR:"]/following-sibling::td/text()')[0]
            try:
                film['film']['cast'] = page.xpath('.//th[text()="REPARTIMENT:"]/following-sibling::td/text()')[0]
            except: 
                continue
            try: 
                film['film']['original_language'] = page.xpath('.//th[text()="IDIOMA PARLAT:"]/following-sibling::td/text()')[0]
            except: 
                continue 
            film['film']['language'] = time.xpath('.//a[@href="{}"]/ancestor::div[contains(@class, "pelicula")]//span'.format(film['link']))
            film['cinema'] = {}
            film['cinema']['name'] = 'Verdi Barcelona'
            film['cinema']['barri'] = 'Gracia'
            film['cinema']['address'] = 'Carrer de Verdi, 32, 08012 Barcelona, Spain'
            screenings.append(film)
        return screenings
```

```python
# Test  
film_scraper = Film_Scraper('https://barcelona.cines-verdi.com/chavalas-vosi')
test = film_scraper.get_film_info()
```

```python
film_results = []
for film in all_films.values(): 
    film_html = Film_Scraper(film)
    film_info = film_html.get_film_info()
    film_results.append(film_info)
```
