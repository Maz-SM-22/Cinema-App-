```python
import requests
import re
from lxml.html import fromstring
import json 
import os
```

```python
url = 'https://www.phenomena-experience.com/programacion-mensual/todo.html'
base_url = url[0:37]
```

```python
class Film_Scraper: 
    def __init__(self, url): 
        self.url = url 
    def get_html(self): 
        try: 
            response = requests.get(self.url)
            response.raise_for_status()
        except requests.RequestException: 
            print('A problem occurred with your request')
            raise
        html = fromstring(response.content)
        return html
    def get_film_info(self):
        page = self.get_html()
        films = page.findall('.//div[@class="event-content"]')
        screenings = []
        film = {}                   # Add genre 
        for block in films: 
            film['link'] = base_url + block.xpath('.//a/@href')[0]
            film['time'] = block.xpath('.//div[@class="event-entrada-hora"]/text()')[0].strip().replace('h', '')
            date = block.xpath('.//a/preceding::div[contains(@class, "dia-titulo")]/text()')[0].strip()
            film['date'] = '2021-10-{}'.format(date.split()[1]) 
            film['film'] = {}
            film['film']['title'] = block.xpath('.//div[@class="event-titulo"]//span/text()')
            film['film']['poster'] = base_url + block.xpath('.//div[@class="event-img"]//img/@src')[0]
            film_info = block.xpath('.//div[@class="event-datos"]/text()')[0].split('Â·')
            film['film']['year'] = film_info[1]
            film['film']['duration'] = film_info[-1] + 'mins'
            film['film']['director'] = film_info[0]
#            film['film']['cast'] = block.xpath('.//div[@class="event-interpretes"]/text()')[0].split()
            film['film']['original_language'] = film_info[2]
            film['film']['language'] = film_info[3]
            film['cinema'] = {}
            film['cinema']['name'] = 'Phenomena'
            film['cinema']['barri'] = 'Eixample'
            film['cinema']['address'] = 'Carrer de Sant Antoni Maria Claret, 168, 08025 Barcelona, Spain'
            screenings.append(film)
        return screenings
```

```python
scraper = Film_Scraper(url)
scraper.get_film_info()
```
